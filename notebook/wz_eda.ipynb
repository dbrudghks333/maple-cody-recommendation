```python
import os
from PIL import Image
```

```python
base_dir = os.path.join('.', '.data')
images = []
for item_type in range(0, 15):
    dir_path = os.path.join(base_dir, f'parts{item_type}')
    item_name_list = sorted(os.listdir(dir_path))
    for item_name in item_name_list:
        item_path = os.path.join(dir_path, item_name)
        with Image.open(item_path) as im:
            image_pixels = list(im.getdata())
            cnt0, cnt1 = 0, 0
            for pix in image_pixels:
                if pix == (0,0,0,0) or pix == (255, 0, 0, 0):
                    cnt0 += 1
                if pix == (0,0,0,17) or pix == (255,0,0,17):
                    cnt1 += 1
            if len(image_pixels) == cnt0 + cnt1:
                continue
            images.append(im)
                
```
